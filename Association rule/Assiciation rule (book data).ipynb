# 3) Visulize the obtained rules using different plots # (Book data)Â¶

```python
import pandas as pd
import networkx as nx  # Add this line
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Apply the Apriori algorithm
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_threshold)

# Visualize using network plot
G = nx.Graph()
for i, row in rules.iterrows():
    G.add_edge(str(row['antecedents']), str(row['consequents']), weight=row['support'])
```

```python
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='skyblue')
nx.draw_networkx_edges(G, pos, width=2, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(str(i[0]), str(i[1])): f"{i[2]:.2f}" for i in G.edges(data='weight')}, font_color='red')
plt.title('Association Rules Network Plot')
plt.show()
```

```python
# Visualize using bar charts
sns.barplot(x='support', y='antecedents', data=rules, palette='viridis', ci=None)
plt.title('Support of Antecedents')
plt.show()
```

```python
sns.barplot(x='confidence', y='antecedents', data=rules, palette='mako', ci=None)
plt.title('Confidence of Antecedents')
plt.show()
```

```python
sns.barplot(x='lift', y='antecedents', data=rules, palette='rocket', ci=None)
plt.title('Lift of Antecedents')
plt.show()
```
