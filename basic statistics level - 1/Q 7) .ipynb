```python
import pandas as pd
from statistics import mean, median, mode, variance, stdev
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\Pritee\OneDrive\Documents\data science assignments\assignments\basic statistics level 1\Q7.csv")
df
```

```python
# Extract columns
points = df['Points']
score = df['Score']
weight = df['Weigh']
```

```python
# Function to calculate mode (as mode may have multiple values)
def calculate_mode(data):
    try:
        mode_val = mode(data)
        return [mode_val]
    except StatisticsError:
        return None
```

```python
# Calculate Mean, Median, Mode, Variance, Standard Deviation, and Range
```

```python
# for points column 
```

```python
mean_points = mean(points)
median_points = median(points)
mode_points = calculate_mode(points)
variance_points = variance(points)
std_dev_points = stdev(points)
range_points = max(points) - min(points)
```

```python
# Print the results
print("For Points:")
print(f"Mean: {mean_points:.2f}")
print(f"Median: {median_points}")
print(f"Mode: {mode_points}")
print(f"Variance: {variance_points:.2f}")
print(f"Standard Deviation: {std_dev_points:.2f}")
print(f"Range: {range_points}\n")
```
