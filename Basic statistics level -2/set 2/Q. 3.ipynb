# Q3.	If X1 ~ N(μ, σ2) and X2 ~ N(μ, σ2) are iid normal random variables, then what is the difference between 2 X1 and X1 + X2? Discuss both their distributions and parameters

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parameters
mu = 2  # Mean for X1 and X2
sigma = 3  # Standard deviation for X1 and X2
```

```python
# Number of samples
num_samples = 100000
```

```python
# Generate random samples for X1 and X2
X1 = np.random.normal(mu, sigma, num_samples)
X2 = np.random.normal(mu, sigma, num_samples)
```

```python
# Calculate 2*X1 and X1 + X2
twice_X1 = 2 * X1
sum_X1_X2 = X1 + X2
```

```python
# Plot the histograms
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(twice_X1, bins=100, density=True, alpha=0.7, color='b', label='2*X1')
x_range = np.linspace(-10, 30, 1000)
plt.plot(x_range, norm.pdf(x_range, 2*mu, 2*sigma), 'r', linewidth=2)
plt.title('2*X1')
plt.legend()
```

```python
plt.subplot(1, 2, 2)
plt.hist(sum_X1_X2, bins=100, density=True, alpha=0.7, color='g', label='X1 + X2')
x_range = np.linspace(-10, 30, 1000)
plt.plot(x_range, norm.pdf(x_range, 2*mu, 2*sigma**2), 'r', linewidth=2)
plt.title('X1 + X2')
plt.legend()

plt.show()
```

# conclusion is The histograms closely match the expected normal distributions for 2X1 and X1 + X2, confirming the discussed parameters.
